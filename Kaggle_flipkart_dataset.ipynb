```python
#8.making variable for perticular brand
apple = df[df.brand == 'Apple']
```

```python
#9.print all entries of variable apple
apple
```

```python
#10.plot the graph of models of apple vs their sales price
plt.plot(apple.model, apple.sales_price, color = 'r')
plt.xlabel("Model name", color = 'r', fontsize = 20)
plt.ylabel("Price in INR", color = 'g', fontsize = 20)
plt.xlim([0,4])
plt.title("models of brand apple vs price in INR")
plt.show()
```

```python
#11
plt.bar(apple.model, apple.ROM, width = 0.5, color = 'r')
plt.xticks(rotation = 45)
plt.show()
```

```python
#16
grouped_color =  df.groupby(['base_color'],as_index=False).model.count()
from matplotlib.pyplot import figure
figure(figsize=(15, 8), dpi=80)
plt.bar(grouped_color['base_color'], grouped_color['model'], color = ['black','blue','#CD7F32','#FFD700','gray', 'green','pink','purple','red','silver','brown','yellow'])
```

```python
#17
grouped_brand =  df.groupby(['brand'],as_index=False).model.count()
figure(figsize=(9, 8), dpi=80)
plt.bar(grouped_brand['brand'], grouped_brand['model'], color = '#e75480')
```
